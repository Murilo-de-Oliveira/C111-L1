```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Plotando gr√°ficos com plot

```python
x = np.array([1,2,3,4])

y = x*2

plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')

plt.plot(x,y)
```

## Formatando estilo dos plots

```python
plt.plot(x, y, 'o:g', linewidth=3, markersize=20)
```

## Plots m√∫ltiplos de forma conjunta

```python
x = np.array([1,2,3,4])

y1 = x*2
y2 = x**2

plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')

plt.plot(x,y1,'r-',x,y2,'b--')
```

## Plots m√∫ltiplos de forma separada

```python
x = np.array([1, 2, 3, 4])
y = x*2
y2 = x*x

plt.subplot(1, 2, 1) #uma linha, duas colunas, posi√ß√£o 1
plt.title('Linear')
plt.plot(x, y, 'r-')
plt.subplot(1, 2, 2) #uma linha, duas colunas, posi√ß√£o 2
plt.title('Exponencial')
plt.plot(x, y2, 'b--')
```

# Tipos de gr√°ficos

## Gr√°fico de dispers√£o

```python
dfPaises = pd.read_csv('paises.csv', delimiter=';')
# extraindo somente dados dos 6 maiores pa√≠ses do mundo
dfPaises2 = dfPaises.nlargest(6, 'Area (sq. mi.)')
# plotando qual destes pa√≠ses possui a maior renda per capita
# observe que o tamanho de cada ponto ilustra o tamanho dos pa√≠ses
plt.scatter(dfPaises2['Country'], dfPaises2['GDP ($ per capita)'],
s=dfPaises2['Area (sq. mi.)']/10000)
```

## Gr√°fico em barras

```python
# extraindo os 5 pa√≠ses com maior PIB per capita (GDP) do dataset
dfBiggestGDP = dfPaises.nlargest(5, 'GDP ($ per capita)')
# pegando as Series dos pa√≠ses e os valores de GDP separadas
dfBiggestGDP_country = dfBiggestGDP['Country']
dfBiggestGDP_gdp = dfBiggestGDP['GDP ($ per capita)']
# tra√ßando um gr√°fico em barras para ilustrar as diferen√ßas
plt.bar(dfBiggestGDP_country, dfBiggestGDP_gdp, color='green')
```

## Gr√°fico em Torta (Pie Plot)

```python
#extraindo os pa√≠ses que n√£o tem costa mar√≠tma
dfPaisesNoCoast = dfPaises[dfPaises['Coastline (coast/area ratio)'] == 0]
#extraindo a quantidade de pa√≠ses que n√£o tem costa mar√≠tima
qtPaisesNoCoast = len(dfPaisesNoCoast)
#extraindo a quantidade de paises que tem costa mar√≠tma
qtPaisesCoast = len(dfPaises) - qtPaisesNoCoast
#tra√ßando o gr√°fico em torta
plt.pie(x=[qtPaisesCoast, qtPaisesNoCoast], labels=['% Pa√≠ses Com Costa',
'% Pa√≠ses Sem Costa'], autopct='%1.1f%%')
```
