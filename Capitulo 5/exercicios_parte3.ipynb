```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('paises.csv',delimiter=';')
print(dataset)
```

1 - Carregue o Dataset paises.csv e em seguida mostre:
a.Quais são os países da OCEANIA;
b.Quantos países são da OCEANIA;

```python
filtro = dataset['Region'].str.contains('OCEANIA')
paises_oceania = dataset[filtro]['Country']
total_paises_oceania = paises_oceania.count()
print(f'Países da Oceania: ')
for paises in paises_oceania:
    print(f' {paises}')
print(f'Número de países da Oceania: {total_paises_oceania}')
```

2 - Encontre o nome e a região do país que possui a maior população segundo este
Dataset

```python
index_maior_populacao = np.argmax(dataset['Population'])
pais_regiao = dataset.loc[[index_maior_populacao],['Country','Region']]
print(f"País com a maior população: {pais_regiao['Country'].values[0]}")
print(f"Região do país com a maior população: {pais_regiao['Region'].values[0]}")
```

3 - Agrupe os países por Regiões. Em seguida, mostre a média de alfabetização
(Literacy (%)) de cada região do planeta

```python
regioes = dataset.groupby('Region')
media_alfabetizacao = regioes['Literacy (%)'].mean()
print(f'Média de alfabetização de cada região: ')
for regiao, media in media_alfabetizacao.items():
    print(f'Região: {regiao} \nMédia de alfabetização: {media:.2f}%\n')
```

4 - Busque o nome de todos os países do Dataset que não possuem costa marítima
(Coastline (coast/area ratio) == 0) e guarde-os em um novo arquivo chamado
noCoast.csv

```python
filtro = dataset['Coastline (coast/area ratio)'] == 0
print(dataset[filtro])
dataset[filtro].to_csv('noCoast.csv',sep=';')
```

5 - Faça uma função que receba a taxa de mortalidade de cada país (Deathrate) e
retorne o texto ‘Balanced’ caso o valor seja < 9 e ‘Urgent’ caso contrário. Em
seguida, crie um campo no Dataset chamado ‘Humanitarian Help’ que receba estes
valores para cada país. No final, mostre o Dataset para verificar se a inserção da nova
coluna foi feita com sucesso

```python
def deathRate(valor):
    if valor >= 9:
        return 'Urgent'
    else:
        return 'Balanced'
```

```python
dataset['Humanitarian Help'] = dataset['Deathrate'].apply(deathRate)
dataset
```
